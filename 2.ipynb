```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef, confusion_matrix
```

```python
wine = load_wine()
```

```python
x = wine.data
x
```

```python
y = wine.target
y
```

```python
dff = pd.DataFrame(x,columns=wine.feature_names)
dff
```

```python
dft = pd.DataFrame(y,columns=['target'])
dft
```

```python
df = pd.concat([dff,dft],axis=1)
df
```

```python
X = df.drop('target',axis=1)
xtr,xte,ytr,yte = train_test_split(X,df['target'],test_size=0.2,random_state=42)
```

```python
xtr,xte,ytr,yte
```

```python
model = DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(xtr,ytr)
```

```python
pred = model.predict(xte)
pred
```

```python
accuracy_score(yte,pred)
```

```python
precision_score(yte,pred,average='macro')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cf = confusion_matrix(yte,pred)
cf
sns.heatmap(cf,annot=True, fmt='d', cmap='Blues', cbar=False, xticklabels=['Class 0', 'Class 1', 'Class 2'], yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
print(cf)
```

```python
from sklearn.metrics import classification_report
print(classification_report(yte,pred))
```

```python
plot_tree(model)
```

```python
recall_score(yte,pred,average='macro')
```
